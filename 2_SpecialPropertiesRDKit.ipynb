# Compute special and derived properties in RDKit

RDKit has a number of functions that enables users to calculate special properties such as the synthetic accessbility (SA) score, NIBR filters and PAINS.

In this notebook, I will document all of the above and others that I might find useful in the future

```python
from rdkit import Chem
from rdkit.Chem import Draw
```

#### The SA score

```python
q= Chem.MolFromSmiles("Fc8cccc(c8)CNC(=O)C3CCN(C(c1c2ccccc2ccc1)C)CC3")
Draw.MolToImage(q)
```

```python
#SAscore is stored in the contrib folder and hence we import it: rdkit_venv/lib/python3.12/site-packages/rdkit/Contrib/SA_Score
#The SA_Score ranges from 1 to 10 with 1 being easy to make and 10 being hard to make.

from rdkit.Contrib.SA_Score import sascorer

print(f"The SA score is {sascorer.calculateScore(q)}")
```
